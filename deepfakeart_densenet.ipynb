```python
# =============================================================================
# Siamese Network for AI Art Forgery Detection (DenseNet DDP Version)
# =============================================================================

import os
import json
import random
from collections import defaultdict
from PIL import Image
import numpy as np

import torch
import torch.nn as nn
import torch.nn.functional as F 
import torch.optim as optim
# IMPORTS FOR DDP
import torch.distributed as dist 
from torch.nn.parallel import DistributedDataParallel as DDP
from torch.utils.data.distributed import DistributedSampler 

from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, models

import matplotlib.pyplot as plt
from tqdm.auto import tqdm

# =============================================================================
# 1. Configuration and DDP Setup
# =============================================================================
# --- Paths ---
BASE_PATH = '/kaggle/input/deepfakeart/'
SIMILAR_PATH = os.path.join(BASE_PATH, 'similar')
DISSIMILAR_PATH = os.path.join(BASE_PATH, 'dissimilar')
OUTPUT_PATH = '/kaggle/working/'

# --- Training Hyperparameters ---
NUM_EPOCHS = 25
BATCH_SIZE = 32 # Total batch size. Each GPU will get BATCH_SIZE / WORLD_SIZE.
LEARNING_RATE = 0.001 
NUM_WORKERS = 4 

# --- DDP Setup ---
WORLD_SIZE = int(os.environ.get('WORLD_SIZE', 1))
RANK = int(os.environ.get('RANK', 0))
LOCAL_RANK = int(os.environ.get('LOCAL_RANK', 0))

def setup_ddp(rank, world_size):
    """Initialize the distributed process group."""
    if world_size == 1: return False
    
    os.environ['MASTER_ADDR'] = os.environ.get('MASTER_ADDR', 'localhost')
    os.environ['MASTER_PORT'] = os.environ.get('MASTER_PORT', '12355')
    dist.init_process_group(backend='nccl', init_method='env://', world_size=world_size, rank=rank)
    torch.cuda.set_device(rank)
    return True

IS_DDP_INITIALIZED = setup_ddp(RANK, WORLD_SIZE)
DEVICE = torch.device(f"cuda:{LOCAL_RANK}" if torch.cuda.is_available() else "cpu")

if RANK == 0:
    print(f"Using device: {DEVICE}")
    print(f"World Size: {WORLD_SIZE}, Local Rank: {LOCAL_RANK}")
    print(f"Number of data loading workers: {NUM_WORKERS}")

# --- Image Transformations (No Change) ---
IMG_SIZE = 224
train_transforms = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(15),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

val_transforms = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# =============================================================================
# 2. Data Loading and Preparation
# =============================================================================
def load_image_pairs_from_json(json_path, image_root_path, label, max_pairs_per_category=None):
    """Loads image pairs from a JSON file, correcting for path issues."""
    pairs = []
    if not os.path.exists(json_path):
        if RANK == 0: print(f"Error: JSON file not found at {json_path}"); return pairs
    
    with open(json_path, 'r') as f: data = json.load(f)
    category_counts = defaultdict(int)

    for category, items in data.items():
        current_category_pairs = []
        for key, value in items.items():
            if 'original' in value and 'generated' in value:
                img1_path = os.path.join(image_root_path, value['original'])
                img2_path = os.path.join(image_root_path, value['generated'])
            elif 'image_0' in value and 'image_1' in value:
                img1_path = os.path.join(image_root_path, value['image_0'])
                img2_path = os.path.join(image_root_path, value['image_1'])
            else: continue

            if os.path.exists(img1_path) and os.path.exists(img2_path):
                current_category_pairs.append((img1_path, img2_path, label))
        
        random.shuffle(current_category_pairs)
        if max_pairs_per_category and len(current_category_pairs) > max_pairs_per_category:
            pairs.extend(current_category_pairs[:max_pairs_per_category])
            category_counts[category] = max_pairs_per_category
        else:
            pairs.extend(current_category_pairs)
            category_counts[category] = len(current_category_pairs)
    
    if RANK == 0:
        print(f"Loaded from {os.path.basename(json_path)}:")
        total = sum(category_counts.values())
        if total > 0:
            for cat, count in category_counts.items():
                print(f"  - Category '{cat}': {count} pairs ({count/total:.2%})")
        print(f"Total pairs: {total}\n")

    return pairs

if RANK == 0: print("--- Loading Training Data ---")
train_similar_pairs = load_image_pairs_from_json(os.path.join(SIMILAR_PATH, 'train_similar.json'), BASE_PATH, 1)
train_dissimilar_pairs = load_image_pairs_from_json(os.path.join(DISSIMILAR_PATH, 'train_dissimilar.json'), BASE_PATH, 0)

if RANK == 0: print("--- Loading Validation Data ---")
test_similar_pairs = load_image_pairs_from_json(os.path.join(SIMILAR_PATH, 'test_similar.json'), BASE_PATH, 1)
test_dissimilar_pairs = load_image_pairs_from_json(os.path.join(DISSIMILAR_PATH, 'test_dissimilar.json'), BASE_PATH, 0)

train_data = train_similar_pairs + train_dissimilar_pairs
val_data = test_similar_pairs + test_dissimilar_pairs
random.shuffle(train_data)
random.shuffle(val_data)

# Store counts for loss weighting (Used in Section 4)
NUM_SIMILAR_TRAIN = len(train_similar_pairs) # 9249
NUM_DISSIMILAR_TRAIN = len(train_dissimilar_pairs) # 16378

if RANK == 0:
    print(f"Total training pairs: {len(train_data)} (Similar: {NUM_SIMILAR_TRAIN}, Dissimilar: {NUM_DISSIMILAR_TRAIN})")
    print(f"Total validation pairs: {len(val_data)} (Similar: {len(test_similar_pairs)}, Dissimilar: {len(test_dissimilar_pairs)})")


# =============================================================================
# 3. Custom Dataset Class and DataLoaders
# =============================================================================
class SiameseDataset(Dataset):
    def __init__(self, image_pairs, transform=None):
        self.image_pairs = image_pairs
        self.transform = transform

    def __len__(self): return len(self.image_pairs)

    def __getitem__(self, index):
        img1_path, img2_path, label = self.image_pairs[index]
        try:
            img1 = Image.open(img1_path).convert('RGB')
            img2 = Image.open(img2_path).convert('RGB')
        except (IOError, FileNotFoundError) as e:
            return None 

        if self.transform:
            img1 = self.transform(img1)
            img2 = self.transform(img2)
            
        return img1, img2, torch.tensor(label, dtype=torch.float32)

def collate_fn(batch):
    batch = list(filter(lambda x: x is not None, batch))
    if len(batch) == 0: return torch.Tensor(), torch.Tensor(), torch.Tensor()
    return torch.utils.data.dataloader.default_collate(batch)

train_dataset = SiameseDataset(train_data, transform=train_transforms)
val_dataset = SiameseDataset(val_data, transform=val_transforms)

# DDP DataLoaders setup
if IS_DDP_INITIALIZED:
    train_sampler = DistributedSampler(train_dataset, num_replicas=WORLD_SIZE, rank=RANK, shuffle=True)
    val_sampler = DistributedSampler(val_dataset, num_replicas=WORLD_SIZE, rank=RANK, shuffle=False)
    
    train_loader = DataLoader(
        dataset=train_dataset, batch_size=BATCH_SIZE // WORLD_SIZE, shuffle=False,
        num_workers=NUM_WORKERS, pin_memory=True, collate_fn=collate_fn, 
        sampler=train_sampler, persistent_workers=True
    )
    val_loader = DataLoader(
        dataset=val_dataset, batch_size=BATCH_SIZE // WORLD_SIZE, shuffle=False,
        num_workers=NUM_WORKERS, pin_memory=True, collate_fn=collate_fn, 
        sampler=val_sampler, persistent_workers=True
    )
else:
    train_loader = DataLoader(
        dataset=train_dataset, batch_size=BATCH_SIZE, shuffle=True,
        num_workers=NUM_WORKERS, pin_memory=True, collate_fn=collate_fn, persistent_workers=True
    )
    val_loader = DataLoader(
        dataset=val_dataset, batch_size=BATCH_SIZE, shuffle=False,
        num_workers=NUM_WORKERS, pin_memory=True, collate_fn=collate_fn, persistent_workers=True
    )


# =============================================================================
# 4. DenseNet-based Siamese Network Model (CRITICAL FIXES APPLIED)
# =============================================================================
class DenseNetEncoder(nn.Module):
    """Encoder using a pre-trained DenseNet-121 backbone."""
    def __init__(self, embedding_dim=512, pretrained=True):
        super(DenseNetEncoder, self).__init__()
        weights = models.DenseNet121_Weights.IMAGENET1K_V1 if pretrained else None
        self.densenet = models.densenet121(weights=weights)
        
        # Extract features and replace the final classification layer
        IN_FEATURES = self.densenet.classifier.in_features
        self.densenet.classifier = nn.Identity()
        
        self.fc = nn.Linear(IN_FEATURES, embedding_dim)
        self.bn = nn.BatchNorm1d(embedding_dim)

    def forward(self, x):
        # DenseNet feature extraction
        x = self.densenet.features(x)
        x = F.relu(x, inplace=True)
        # Global Average Pooling (converts 4D tensor to 2D [Batch, Channels])
        x = F.adaptive_avg_pool2d(x, (1, 1)).view(x.size(0), -1) 
        
        # Final embedding layers
        x = self.fc(x)
        x = self.bn(x)
        return x

class SiameseNetwork(nn.Module):
    """
    Siamese Network with a Prediction Head for correct logit output.
    """
    def __init__(self, encoder):
        super(SiameseNetwork, self).__init__()
        self.encoder = encoder
        self.similarity = nn.CosineSimilarity(dim=1, eps=1e-6)
        
        #  FIX: Prediction head to convert scalar similarity score 
        # (range -1 to 1) into a true logit (range -inf to +inf)
        self.prediction_head = nn.Sequential(
            nn.Linear(1, 1),
        )

    def forward(self, img1, img2):
        emb1 = self.encoder(img1)
        emb2 = self.encoder(img2)
        
        sim_score = self.similarity(emb1, emb2) 
        
        # Reshape to [Batch_Size, 1] for the Linear layer input
        sim_score = sim_score.unsqueeze(1) 
        
        # Output the raw logit score expected by nn.BCEWithLogitsLoss
        logit = self.prediction_head(sim_score).squeeze(1)
        
        return logit 

# --- Model Instantiation and Loss Weighting ---
encoder = DenseNetEncoder().to(DEVICE)
model = SiameseNetwork(encoder).to(DEVICE)

#  FIX: Class Imbalance Correction (pos_weight = 16378 / 9249  1.7708)
POS_WEIGHT = torch.tensor(NUM_DISSIMILAR_TRAIN / NUM_SIMILAR_TRAIN, dtype=torch.float32)

if IS_DDP_INITIALIZED:
    model = DDP(model, device_ids=[LOCAL_RANK]) 
    if RANK == 0:
        print(f"Applying Class Weighting: pos_weight = {POS_WEIGHT.item():.4f}")

# CRITICAL: Apply pos_weight to the loss function
criterion = nn.BCEWithLogitsLoss(pos_weight=POS_WEIGHT.to(DEVICE)) 
optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='max', factor=0.5, patience=2, verbose=False)

# =============================================================================
# 5. Training and Validation Loop (DDP Compatible)
# =============================================================================
def train_one_epoch(model, loader, criterion, optimizer, device, epoch):
    model.train()
    running_loss, correct_predictions, total_samples = 0.0, 0, 0
    
    if IS_DDP_INITIALIZED:
        loader.sampler.set_epoch(epoch) 

    progress_bar = tqdm(loader, desc=f"Training Epoch {epoch}", leave=False, disable=(RANK != 0))
    
    for img1, img2, labels in progress_bar:
        if img1.nelement() == 0: continue
        img1, img2, labels = img1.to(device), img2.to(device), labels.to(device)
        
        optimizer.zero_grad()
        outputs = model(img1, img2)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item() * img1.size(0)
        preds = (outputs > 0).float()
        correct_predictions += (preds == labels).sum().item()
        total_samples += labels.size(0)
        
        if RANK == 0:
             progress_bar.set_postfix(
                loss=f"{running_loss/total_samples:.4f}", 
                acc=f"{correct_predictions/total_samples:.4f}"
            )
    
    return running_loss / total_samples, correct_predictions / total_samples

def validate_one_epoch(model, loader, criterion, device):
    model.eval()
    running_loss, correct_predictions, total_samples = 0.0, 0, 0
    
    progress_bar = tqdm(loader, desc="Validating", leave=False, disable=(RANK != 0))
    
    with torch.no_grad():
        for img1, img2, labels in progress_bar:
            if img1.nelement() == 0: continue
            img1, img2, labels = img1.to(device), img2.to(device), labels.to(device)
            
            outputs = model(img1, img2)
            loss = criterion(outputs, labels)
            
            running_loss += loss.item() * img1.size(0)
            preds = (outputs > 0).float()
            correct_predictions += (preds == labels).sum().item()
            total_samples += labels.size(0)
            
            if RANK == 0:
                progress_bar.set_postfix(
                    loss=f"{running_loss/total_samples:.4f}", 
                    acc=f"{correct_predictions/total_samples:.4f}"
                )
    
    return running_loss / total_samples, correct_predictions / total_samples

# --- Main Training Loop ---
best_val_acc = 0.0
history = {'train_loss': [], 'train_acc': [], 'val_loss': [], 'val_acc': []}

if RANK == 0: print("\n--- Starting DenseNet DDP Training ---")
for epoch in range(NUM_EPOCHS):
    train_loss, train_acc = train_one_epoch(model, train_loader, criterion, optimizer, DEVICE, epoch)
    val_loss, val_acc = validate_one_epoch(model, val_loader, criterion, DEVICE)
    
    if RANK == 0:
        history['train_loss'].append(train_loss); history['train_acc'].append(train_acc)
        history['val_loss'].append(val_loss); history['val_acc'].append(val_acc)
        
        scheduler.step(val_acc)
        
        print(f"Epoch {epoch+1}/{NUM_EPOCHS} | "
              f"Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.4f} | "
              f"Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.4f}")

        # Saving Model Logic (Only save from Rank 0)
        if val_acc > best_val_acc:
            best_val_acc = val_acc
            # When saving, unwrap the model from DDP
            state_dict = model.module.state_dict() if IS_DDP_INITIALIZED else model.state_dict()
            encoder_state_dict = model.module.encoder.state_dict() if IS_DDP_INITIALIZED else model.encoder.state_dict()
            
            checkpoint_name = f"densenet_ddp_model_acc_{val_acc:.4f}_epoch_{epoch+1}.pth"
            weights_name = f"densenet_ddp_encoder_acc_{val_acc:.4f}_epoch_{epoch+1}.pth"
            torch.save(state_dict, os.path.join(OUTPUT_PATH, checkpoint_name))
            torch.save(encoder_state_dict, os.path.join(OUTPUT_PATH, weights_name))
            print(f"  -> New best DDP DenseNet model saved with validation accuracy: {val_acc:.4f}")

if RANK == 0:
    print("\n--- Training Finished ---")
    # Saving Final Model (Unwrap DDP)
    state_dict = model.module.state_dict() if IS_DDP_INITIALIZED else model.state_dict()
    encoder_state_dict = model.module.encoder.state_dict() if IS_DDP_INITIALIZED else model.encoder.state_dict()
    
    torch.save(state_dict, os.path.join(OUTPUT_PATH, 'siamese_densenet_ddp_full_model_final.pth'))
    torch.save(encoder_state_dict, os.path.join(OUTPUT_PATH, 'densenet_ddp_encoder_final.pth'))
    print(f"Final DenseNet model and encoder weights saved to {OUTPUT_PATH}")

# =============================================================================
# 6. Results Visualization (Only run on Rank 0)
# =============================================================================
if RANK == 0:
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1); plt.plot(history['train_loss'], label='Train Loss'); plt.plot(history['val_loss'], label='Validation Loss')
    plt.title('Loss Over Epochs (DenseNet DDP)'); plt.xlabel('Epochs'); plt.ylabel('Loss'); plt.legend()
    plt.subplot(1, 2, 2); plt.plot(history['train_acc'], label='Train Accuracy'); plt.plot(history['val_acc'], label='Validation Accuracy')
    plt.title('Accuracy Over Epochs (DenseNet DDP)'); plt.xlabel('Epochs'); plt.ylabel('Accuracy'); plt.legend()
    plt.tight_layout()
    plt.savefig(os.path.join(OUTPUT_PATH, 'densenet_ddp_training_history.png'))
    plt.show()

# =============================================================================
# 7. Inference Function 
# =============================================================================
def predict_similarity(image_path1, image_path2, model_path, device, transform):
    encoder = DenseNetEncoder() # Use the DenseNet encoder
    model = SiameseNetwork(encoder)
    model.load_state_dict(torch.load(model_path, map_location=device))
    model.to(device); model.eval()
    
    try:
        img1 = Image.open(image_path1).convert('RGB'); img2 = Image.open(image_path2).convert('RGB')
    except (IOError, FileNotFoundError) as e:
        print(f"Error loading images for inference: {e}")
        return None, "Error"
    
    img1_t = transform(img1).unsqueeze(0).to(device); img2_t = transform(img2).unsqueeze(0).to(device)
    
    with torch.no_grad():
        logit = model(img1_t, img2_t).item()
        # Convert logit back to similarity probability (0 to 1) for interpretation
        similarity = torch.sigmoid(torch.tensor(logit)).item() 
        
    return similarity, 'Similar/Forgery' if similarity > 0.5 else 'Dissimilar'

# =============================================================================
# 8. Example Inference Test (Only run on Rank 0)
# =============================================================================
if RANK == 0:
    print("\n--- Running Inference Test ---")
    saved_model_path = os.path.join(OUTPUT_PATH, 'siamese_densenet_ddp_full_model_final.pth')

    if os.path.exists(saved_model_path) and val_data:
        test_similar_pair = next((p for p in val_data if p[2] == 1), None)
        test_dissimilar_pair = next((p for p in val_data if p[2] == 0), None)
        
        # Test Similar Pair
        if test_similar_pair:
            img1_path, img2_path, _ = test_similar_pair
            sim_score, pred_label = predict_similarity(img1_path, img2_path, saved_model_path, DEVICE, val_transforms)
            print(f"Test on SIMILAR pair:\n  -> Pred: {pred_label} (Score: {sim_score:.4f}) | GT: Similar/Forgery")
            
            fig, axs = plt.subplots(1, 2, figsize=(8, 4)); axs[0].imshow(Image.open(img1_path)); axs[0].set_title('Image 1'); axs[0].axis('off')
            axs[1].imshow(Image.open(img2_path)); axs[1].set_title('Image 2'); axs[1].axis('off')
            plt.suptitle(f'Similar Pair Test - Prediction: {pred_label}'); plt.show()

        # Test Dissimilar Pair
        if test_dissimilar_pair:
            img1_path, img2_path, _ = test_dissimilar_pair
            sim_score, pred_label = predict_similarity(img1_path, img2_path, saved_model_path, DEVICE, val_transforms)
            print(f"\nTest on DISSIMILAR pair:\n  -> Pred: {pred_label} (Score: {sim_score:.4f}) | GT: Dissimilar")

            fig, axs = plt.subplots(1, 2, figsize=(8, 4)); axs[0].imshow(Image.open(img1_path)); axs[0].set_title('Image 1'); axs[0].axis('off')
            axs[1].imshow(Image.open(img2_path)); axs[1].set_title('Image 2'); axs[1].axis('off')
            plt.suptitle(f'Dissimilar Pair Test - Prediction: {pred_label}'); plt.show()
    else:
        print("Could not find saved model or validation data for inference test.")

# =============================================================================
# 9. Cleanup
# =============================================================================
if IS_DDP_INITIALIZED:
    dist.destroy_process_group()
```
